```python
# Define the report dates
report_dates = {
    "RosterReport_Jan2023": "2023-01-01",
    "RosterReport_March2023": "2023-03-31",
    "RosterReport_June2023": "2023-06-30",
    "RosterReport_Sept2023": "2023-09-30",
    "RosterReport_Dec2023": "2023-12-31"
}

# Convert the date strings to datetime objects
report_dates = {k: datetime.strptime(v, "%Y-%m-%d") for k, v in report_dates.items()}

# Add the "Report Month" column to each DataFrame
RosterReport_Jan2023["Report Month"] = report_dates["RosterReport_Jan2023"]
RosterReport_March2023["Report Month"] = report_dates["RosterReport_March2023"]
RosterReport_June2023["Report Month"] = report_dates["RosterReport_June2023"]
RosterReport_Sept2023["Report Month"] = report_dates["RosterReport_Sept2023"]
RosterReport_Dec2023["Report Month"] = report_dates["RosterReport_Dec2023"]
```

```python
Reports = [RosterReport_Sept2023,RosterReport_Dec2023,RosterReport_Jan2023,RosterReport_June2023,RosterReport_March2023]
Cleaned_Reports = []
def pre_cleaning (Reports):
    for Report in Reports:  
        #cleaning the necessary columns
        Report = Report.rename(columns={"employee_id":"Employee ID", "hire_date":"Hire Date", 
                                                                    "employee_status":"Employee Status",
                                                                    "employee_type":"Employee Type", "job_level":"Job Level", 
                                                                    "supervisor_status": "Supervisor Status","union_affiliation":"Union_Affiliation", 
                                                                    "age_group":"Age Group"})
        Report['Employee Type'] = Report['Employee Type'].replace({'contract': 'Contract', 'part-time': 'Part-Time', 'full-time': 'Full-Time'})
        Report['Employee Status'] = Report['Employee Status'].replace({'on leave': 'On Leave', 'active': 'Active'})
        Report['Union_Affiliation'] = Report['Union_Affiliation'].replace({'Non-Uion': 'Non-Union'})
        Report['Job Level'] = Report['Job Level'].replace({'management': 'Management', 'executive': 'Executive', 'staff': 'Staff'})

        Report['Hire Date'] = pd.to_datetime(Report['Hire Date'])
        today = pd.to_datetime(datetime.now().date())
# Calculate the difference between hire date and today and add it to the new column created
        Report['Service Time (Days)'] = (today - Report['Hire Date']).dt.days
        Report['Service Time'] = Report['Service Time (Days)'] / 365.25
        Report.drop(columns=['Service Time (Days)'], inplace=True)

        Cleaned_Reports.append(Report)
    return Cleaned_Reports
```

```python
Cleaned_Reports = pre_cleaning(Reports)
```
