```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
Terminations2023 = pd.read_csv('./Terminations2023MOCK_DATA.csv')
```

```python
#Looking at the column names
Terminations2023.columns
```

```python
#cleaning the necessary columns
Terminations2023_Cleaned = Terminations2023.rename(columns={"employee_id":"Employee ID", "hire_date":"Hire Date", 
                                                                    "employee_status":"Employee Status",
                                                                    "employee_type":"Employee Type", "job_level":"Job Level", 
                                                                    "supervisor_status": "Supervisor Status","union_affiliation":"Union_Affiliation", 
                                                                    "age_group":"Age Group",
                                                                    #Termination file specific columns
                                                                    "termination_date":"Termination Date", "termination_type": "Termination Type", 
                                                                    "termination_reason":"Termination Reason", "last performance rating":"Last Performance Rating",
                                                                    "exit interview completed":"Exit Interview Completed","new company":"New Company", "rehire eligibility": "Rehire Eligibility"})
#confirming the columm cleaning worked 
Terminations2023_Cleaned.columns
```

```python
#Checking the data
Terminations2023.head(10)
```

```python
Terminations2023_Cleaned['Termination Reason'].value_counts()
```

```python
Terminations2023_Cleaned['Employee Type'] = Terminations2023_Cleaned['Employee Type'].replace({'contract': 'Contract', 'part-time': 'Part-Time', 'full-time': 'Full-Time'})
Terminations2023_Cleaned['Employee Status'] = Terminations2023_Cleaned['Employee Status'].replace({'terminated': 'Terminated'})
Terminations2023_Cleaned['Union_Affiliation'] = Terminations2023_Cleaned['Union_Affiliation'].replace({'Non-Uion': 'Non-Union'})
Terminations2023_Cleaned['Job Level'] = Terminations2023_Cleaned['Job Level'].replace({'management': 'Management', 'executive': 'Executive', 'staff': 'Staff'})

#Promotion File specific columns that require cleaning
Terminations2023_Cleaned['Termination Type'] = Terminations2023_Cleaned['Termination Type'].replace({'retirement': 'Retirement', "voluntary":"Voluntary","involuntary":"Involuntary"})
Terminations2023_Cleaned['Termination Reason'] = Terminations2023_Cleaned['Termination Reason'].replace({'retirement': 'Retirement', "commute to work":"Commute to Work","relocation":"Relocation","company restructuring":"Company Restructuring","poor performance":"Poor Performance",
                                                                                                         "manager":"Manager","better opportunity":"Better Opportunity","career change":"Career Change","personal reasons":"Personal Reasons","attendance":"Attendance","misconduct":"Misconduct"})
Terminations2023_Cleaned.head(20)
```
