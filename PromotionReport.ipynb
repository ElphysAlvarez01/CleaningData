The following columns that the ones requiring formating editing

```python
#cleaning the necessary columns
Promotions2023_Cleaned = Promotions2023.rename(columns={"employee_id":"Employee ID", "hire_date":"Hire Date", 
                                                                    "employee_status":"Employee Status",
                                                                    "employee_type":"Employee Type", "job_level":"Job Level", 
                                                                    "supervisor_status": "Supervisor Status","union_affiliation":"Union_Affiliation", 
                                                                    "age_group":"Age Group",
                                                                    
                                                                    "promotion_date":"Promotion Date", "last promotion date (if applicable)": "Last Promotion Date", 
                                                                    "promotion_reason":"Promotion Reason", "last_performance_rating":"Last Performance Rating",
                                                                    "2022_midyear_trending_rating":"2022 Mid-Year Trending Rating", "2022_midyear_promotion_nomination":"2022 Mid-Year Promotion Nomination"})
#confirming the columm cleaning worked 
Promotions2023_Cleaned.columns
```

Checking that it worked:

```python
#Checking the data
Promotions2023_Cleaned.head(10)
```

Formating the content for the columns that need it. 

```python
Promotions2023_Cleaned['Employee Type'] = Promotions2023_Cleaned['Employee Type'].replace({'contract': 'Contract', 'part-time': 'Part-Time', 'full-time': 'Full-Time'})
Promotions2023_Cleaned['Employee Status'] = Promotions2023_Cleaned['Employee Status'].replace({'on leave': 'On Leave', 'active': 'Active'})
Promotions2023_Cleaned['Union_Affiliation'] = Promotions2023_Cleaned['Union_Affiliation'].replace({'Non-Uion': 'Non-Union'})
Promotions2023_Cleaned['Job Level'] = Promotions2023_Cleaned['Job Level'].replace({'management': 'Management', 'executive': 'Executive', 'staff': 'Staff'})

#Promotion File specific columns that require cleaning
Promotions2023_Cleaned['Promotion Reason'] = Promotions2023_Cleaned['Promotion Reason'].replace({'promotion within compensation band': 'Promotion within Compensation Band', "promotion with new band":"Promotion with New Band"})
Promotions2023_Cleaned['2022 Mid-Year Promotion Nomination'] = Promotions2023_Cleaned['2022 Mid-Year Promotion Nomination'].replace({'consider for next cycle': 'Consider for Next Cycle', 'not applicable': 'Not Applicable', 'promote now': 'Promote Now'})

Promotions2023_Cleaned.head(20)
```

```python
Promotions2023_Cleaned.columns
```

Calculation: Create New Column called "Time Between Promotion" based on the "Promotion Date" and "Last Promotion". I want to analyze the time between promotion and other variables such as Department and Job Level.

```python

Promotions2023_Cleaned['Promotion Date'] = pd.to_datetime(Promotions2023_Cleaned['Promotion Date'])
Promotions2023_Cleaned['Last Promotion Date'] = pd.to_datetime(Promotions2023_Cleaned['Last Promotion Date'])

# Calculate the difference in days
Promotions2023_Cleaned['Time Between Promotion (Days)'] = (Promotions2023_Cleaned['Promotion Date'] - Promotions2023_Cleaned['Last Promotion Date']).dt.days
# Convert the difference in days to years with decimals
Promotions2023_Cleaned['Time Between Promotion'] = Promotions2023_Cleaned['Time Between Promotion (Days)'] / 365.25
Promotions2023_Cleaned.drop(columns=['Time Between Promotion (Days)'], inplace=True)


Promotions2023_Cleaned.head()
```
