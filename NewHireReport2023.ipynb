```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
NewHireReport2023 = pd.read_csv('./NewHires2023MOCK_DATA.csv')
NewHireReport2023.head(15)
```

```python
NewHireReport2023.columns
```

```python
#cleaning the necessary columns
NewHireReport2023_Cleaned = NewHireReport2023.rename(columns={"employee_id":"Employee ID", "hire_date":"Hire Date", 
                                                                    "employee_status":"Employee Status",
                                                                    "employee_type":"Employee Type", "job_level":"Job Level", 
                                                                    "supervisor_status": "Supervisor Status","union_affiliation":"Union_Affiliation", 
                                                                    "age_group":"Age Group"})
#confirming the columm cleaning worked 
NewHireReport2023_Cleaned.columns
```

```python
NewHireReport2023.head(15)
```

```python
NewHireReport2023_Cleaned['Employee Type'] = NewHireReport2023_Cleaned['Employee Type'].replace({'contract': 'Contract', 'part-time': 'Part-Time', 'full-time': 'Full-Time'})
NewHireReport2023_Cleaned['Employee Status'] = NewHireReport2023_Cleaned['Employee Status'].replace({'on leave': 'On Leave', 'active': 'Active'})
NewHireReport2023_Cleaned['Union_Affiliation'] = NewHireReport2023_Cleaned['Union_Affiliation'].replace({'Non-Uion': 'Non-Union'})
NewHireReport2023_Cleaned['Job Level'] = NewHireReport2023_Cleaned['Job Level'].replace({'management': 'Management', 'executive': 'Executive', 'staff': 'Staff'})

NewHireReport2023_Cleaned.head(20)
```

```python
# Top 10 'Prior Company'
top_prior_company = NewHireReport2023_Cleaned['Prior Company'].value_counts().head(10)
print("Top 10 Prior Company:")
print(top_prior_company)

# Top 10 'Prior School'
top_prior_school = NewHireReport2023_Cleaned['Prior School'].value_counts().head(10)
print("\nTop 10 Prior School:")
print(top_prior_school)

# Results are not useful because there doesnt seen to be a large amount of employee coming from any particular school or company
```

```python
# Export CSV file
NewHireReport2023_Cleaned.to_csv('NewHireReport2023_Cleaned.csv', index=False)
```
